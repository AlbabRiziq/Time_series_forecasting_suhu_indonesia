## MERAMAL KENAIKAN RATA-RATA SUHU DI INDONESIA MENGGUNAKAN TIME SERIES ANALYTICS

######   Peramalan merupakan perkiraan untuk suatu kejadian di waktu yang akan datang. Untuk diperoleh hasil peramalan yang baik perlu tersedia data deret waktu yang memadai dan pemilihan metode peramalan yang tepat sesuai dengan karakteristik dari data deret waktu yang diketahui melalui hasil identifikasi yang dilakukan sebelum dilakukan proses pemodelan dan peramalan. Peramalan yang baik yang salah satunya dicirikan dengan tingkat keakuratan dan kemampuan menirukan perilaku data deret waktu historisnya. 


###### Analisis deret waktu (time series) adalah analisis yang mempertimbangkan pengaruh waktu secara berurutan. Sedangkan data deret waktu itu sendiri adalah data yang dikumpulkan berdasarkan urutan dan interval waktu tertentu, seperti dalam jam, hari, minggu, bulan, kuartal, semester dan tahun.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df = pd.read_csv("dataset.csv")
pd.DataFrame(df)

suhu = pd.DataFrame(df, columns=['Indonesia']).to_numpy().ravel()
tahun = pd.DataFrame(df, columns=['Tahun']).to_numpy().ravel()
panjangData = len(tahun)

```

```python
plt.xlabel("Tahun")
plt.ylabel("Suhu(Â°C)")
plt.plot(tahun, suhu, label="Data suhu")
plt.legend(loc="upper left")
```

### TREND LINE LINEAR REGRESSION


###### Pola data trend(Trend Line) merupakan komponen data deret waktu yang menunjukkan peningkatan atau penurunan dalam jangka panjang selama periode waktu yang akan diamati. Pola data tren terjadi jika terdapat kenaikan atau penurunan sekuler jangka panjang dalam data . Pola data tren menunjukkan pergerakan data yang cenderung meningkat dan menurun dalam waktu yang lama. Sehingga pola data tren merupakan kenaikan dan penurunan data yang terjadi dalam waktu yang lama. Untuk pola data trend dapat dilihat pada grafik dibawah

```python
z = np.polyfit(tahun, suhu, 1)
p = np.poly1d(z)
```

```python
# rata_rata = np.mean(suhu)
plt.title("Suhu Indonesia")
plt.xlabel("Tahun")
plt.ylabel("Suhu(Â°C)")
style.use('seaborn')
plt.scatter(tahun, suhu, label="Data suhu") 
plt.plot(tahun, p(tahun), color='red', label="Trend Line")
plt.legend(loc="upper left")

# 
```
